```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
dat = pd.read_csv("/content/sample_data/metrics/metrics_band10.csv")
dat.columns
```

```python
# The of your CSV-files
csv_files = os.listdir("/content/sample_data/metrics")
```

```python
# New plot
plt.figure()

# Loop over CSV files
for file in csv_files:
    # Read the CSV file into a DataFrame
    df = pd.read_csv("/content/sample_data/metrics/" + file)

    # calculate moving average to plot stepwise progress
    train_loss = df[df['train_loss_step'].notna()]['train_loss_step'].rolling(window=20).mean()

    # Plot the requested column
    plt.plot(df[df['train_loss_step'].notna()]['step'] , train_loss, label=file)

# Add a legend and labels
plt.legend(bbox_to_anchor=(1.05, 0.5), loc='center left')
plt.xlabel('Epoch')
plt.ylabel('Validation precision')
plt.title('Validation Precision over Epochs for different model runs')

# Display the plot
plt.show()
```

```python
# load the csv file
df = pd.read_csv("training_log.csv")

# "train_loss" is the column with the logged losses
# we use the rolling method to apply a moving average
df["train_loss_smoothed"] = df["train_loss"].rolling(window=10).mean()

# plot the smoothed data
plt.figure(figsize=(10,6))
plt.plot(df["epoch"], df["train_loss_smoothed"], label='Smoothed Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

Metrics: Precisition and Recall



Precisionrecall.svg.png

$$
\text{Recall} = \frac{\sum(\text{TP})}{\sum(\text{TP} + \text{FN})}, \text{ Precision} = \frac{\sum(\text{TP})}{\sum(\text{TP} + \text{FP})}
$$

```python
# Create a new plot
plt.figure()

# Loop over your CSV files
for file in csv_files:
    # Read the CSV file into a DataFrame
    df = pd.read_csv("/content/sample_data/metrics/" + file)

    # Plot the 'val_loss' column
    plt.plot(df[df['val_recall'].notna()]['epoch'], df[df['val_recall'].notna()]['val_recall'], label=file)

# Add a legend and labels
plt.legend(bbox_to_anchor=(1.05, 0.5), loc='center left')
plt.xlabel('Epoch')
plt.ylabel('Validation Recall')
plt.title('Validation Recall over Epochs for different bands')

# Display the plot
plt.show()
```
