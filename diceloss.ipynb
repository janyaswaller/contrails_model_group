```python
import pandas as pd
```

```python
masks = pd.read_csv("metrics-mask4epoch.csv")
masks29epochs = pd.read_csv("metrics-mask29epochs.csv")
masksUnetPlusPlus = pd.read_csv("metrics-maskUnetPlusPlus.csv")
Unet = pd.read_csv("metrics-Unet.csv")
Unet2 = pd.read_csv("metricsUnet2.csv")
UnetPLusPLus = pd.read_csv("metricsUnetPlusPlus.csv")
```

```python
masks.dropna(0, subset=["val_dice"], inplace=True)
masks29epochs.dropna(0, subset=["val_dice"], inplace=True)
masksUnetPlusPlus.dropna(0, subset=["val_dice"], inplace=True)
Unet.dropna(0, subset=["val_dice"], inplace=True)
Unet2.dropna(0, subset=["val_dice"], inplace=True)
UnetPLusPLus.dropna(0, subset=["val_dice"], inplace=True)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))

plt.plot(masks['val_dice'], label='masksLinknet4epochs')
plt.plot(masks29epochs['val_dice'], label='masksLinknet29epochs')
plt.plot(masksUnetPlusPlus['val_dice'], label='masksUnetPlusPlus')
plt.plot(Unet['val_dice'], label='Unet')
plt.plot(Unet2['val_dice'], label='Unet2')
plt.plot(UnetPLusPLus['val_dice'], label='UnetPlusPlus')


plt.xlabel('Steps')
plt.ylabel('Validation Dice')
plt.title('Line Plot of validation dice')
plt.legend()
plt.grid(True)
#plt.xlim(0, 10000)
plt.show()
```
